##### 데이터 선택(loc 활용)
+ 이름을 이용하여 원하는 데이터를 선택
+ 데이터.loc[조회할 행의 인덱스, 조회할 컬럼]

```python
data={
    '이름': ['김자바', '이자바', '박자바', '최캐드','정캐드'],
    '학급' : ['자바반', '자바반','자바반', '캐드반','캐드반'],
    '국어점수': [90,80,70,100,90],
    '영어점수':[70,80,80,70,90],
    '평균': [80.0,80.0,85.0,85.0,90.0],
    '취미': ['독서', '','','요가', '축구']
}
```

```python
import pandas as pd
df= pd.DataFrame(data,index=['1번','2번','3번','4번','5번'])
df
```

```python
# 만들어진 dataframe에 index추가
df.index = ['1번','2번','3번','4번','5번']
df
```

```python
# 특정 컬럼값만 조회 
df['이름']
df[['이름','학급']]

# 특정 행의 데이터만 조회(index로)
df.loc['1번']
df.loc[['1번','2번']] 
```

```python
# 특정 행과 컬럼에 대한 데이터 조회
df.loc[['1번','2번'], '이름']
df.loc[['1번','2번'], ['국어점수','영어점수','평균']]
```

```python
# 특정 행과 컬럼에 대한 데이터 조회 두번째
df.loc['1번':'4번','이름':'평균']
```

##### 데이터 선택(i(index)loc 활용)
