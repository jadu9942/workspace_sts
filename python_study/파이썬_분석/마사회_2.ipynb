```python
def makeDic(items, co2_list, dewPoint_list, fireYn_list, mesurDy_list, mesurTime_list, nh3_list, o2_list, placeNm_list, placeSeq_list, reh_list, temper_list) :

    for item in items :
        co2_list.append(item.find('co2').text)
        dewPoint_list.append(item.find('dewPoint').text)
        fireYn_list.append(item.find('fireYn').text)
        mesurDy_list.append(item.find('mesurDy').text)
        mesurTime_list.append(item.find('mesurTime').text)
        nh3_list.append(item.find('nh3').text)
        o2_list.append(item.find('o2').text)
        placeNm_list.append(item.find('placeNm').text)
        placeSeq_list.append(item.find('placeSeq').text)
        reh_list.append(item.find('reh').text)
        temper_list.append(item.find('temper').text)

    
    return co2_list, dewPoint_list, fireYn_list, mesurDy_list, mesurTime_list, nh3_list, o2_list, placeNm_list, placeSeq_list, reh_list, temper_list
```

```python

new_df = pandas.read_excel('DITEST1.xlsx')
new_df
```

```python
#시간대별 일별 오전(07:00~12:00), 오후(12:00~18:00), 야간(18:00:22:00)  판단
type_list = []
type = ''
for mesurTime in new_df['mesurTime']:
    if int(mesurTime) >= 700 and int(mesurTime) < 1200:
        # print(int(mesurTime), '오전')
        type = '오전'
    elif int(mesurTime) >= 1200 and int(mesurTime) < 1800:
        #print(int(mesurTime), '오후')
        type = '오후'
    elif int(mesurTime) >= 1800 and int(mesurTime) < 2200:
        #print(int(mesurTime), '저녁')
        type = '저녁'
    else :
        #print(int(mesurTime), '무엇도 아님') 
        type = ''

    type_list.append(type)
    #print(int(mesurTime))
```

```python
type_list
```

```python
new_df['type'] = type_list
new_df
```

```python
new_df[new_df['type'] == ''].index
```

```python
# 오전, 오후, 저녁을 제외한 데이터는 삭제
new_df.drop(new_df[new_df['type'] == ''].index, inplace=True)
```

```python
level_list = []
level = ''
for DI in new_df['DI']:
    if float(DI) >= 80:
        level = '매우높음'
    elif float(DI) >= 75 :
        level = '높음'
    elif float(DI) >= 68 :        
        level = '보통'
    else :
        level = '낮음'
    level_list.append(level)
new_df['DI_level'] = level_list
new_df
```
